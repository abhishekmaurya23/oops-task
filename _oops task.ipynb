Certainly! Here's an example of a `Vehicle` class with an `__init__` method that initializes instance variables for the name of the vehicle, maximum speed, and average speed:

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_speed):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_speed = average_speed

# Usage
car = Vehicle("Car", 200, 80)
print("Name of Vehicle:", car.name_of_vehicle)
print("Max Speed:", car.max_speed)
print("Average Speed:", car.average_speed)
```

It takes arguments `name_of_vehicle`, `max_speed`, and `average_speed`, representing the properties of the vehicle.

Inside the `__init__` method, the values passed as arguments are assigned to the corresponding instance variables using the `self` keyword. The `self` parameter refers to the instance of the class being created.

After defining the `Vehicle` class, we create an instance of the class named `car` with the values "Car" for `name_of_vehicle`, 200 for `max_speed`, and 80 for `average_speed`. We then access and print the values of the instance variables using dot notation.

Output:
```
Name of Vehicle: Car
Max Speed: 200
Average Speed: 80
```

By providing the necessary arguments while creating an instance of the `Vehicle` class, we can initialize the instance variables and store specific data for each instance of the class.

ans:3-Multiple inheritance is a feature in object-oriented programming where a class can inherit attributes and methods from multiple parent classes. In Python, a class can inherit from multiple base classes, allowing it to incorporate behaviors and characteristics from multiple sources.

Here's an example code that demonstrates multiple inheritance:

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")


class Flyable:
    def fly(self):
        print(f"{self.name} is flying.")


class Bird(Animal, Flyable):
    def __init__(self, name):
        super().__init__(name)

    def chirp(self):
        print(f"{self.name} is chirping.")


# Usage
bird = Bird("Sparrow")
bird.eat()
bird.sleep()
bird.fly()
bird.chirp()
```

In this example, we have three classes: `Animal`, `Flyable`, and `Bird`. The `Animal` class represents common attributes and behaviors of animals, such as eating and sleeping. The `Flyable` class represents the ability to fly. The `Bird` class inherits from both `Animal` and `Flyable` using multiple inheritance.

The `Bird` class overrides the `__init__` method to initialize the `name` attribute by calling the `super().__init__(name)` method. It also defines its specific behavior, `chirp`, which is unique to birds.

By inheriting from both `Animal` and `Flyable`, the `Bird` class incorporates attributes and methods from both parent classes. It can access the `name` attribute and methods like `eat` and `sleep` from the `Animal` class, as well as the `fly` method from the `Flyable` class.

In the usage section, we create an instance of the `Bird` class named `bird` with the name "Sparrow". We can then call methods such as `eat`, `sleep`, `fly`, and `chirp` on the `bird` object.

Output:
```
Sparrow is eating.
Sparrow is sleeping.
Sparrow is flying.
Sparrow is chirping.
```

By using multiple inheritance, we can create classes that combine features and behaviors from multiple parent classes, allowing for greater flexibility and code reuse.

ans:4-In Python, getter and setter methods are used to control the access and modification of class attributes. They provide an interface to retrieve (getter) or modify (setter) the values of private attributes, allowing encapsulation and data protection within a class.

Here's an example that demonstrates the use of getter and setter methods in a class:

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    # Getter method for the name attribute
    def get_name(self):
        return self._name

    # Setter method for the name attribute
    def set_name(self, name):
        self._name = name

    # Getter method for the age attribute
    def get_age(self):
        return self._age

    # Setter method for the age attribute
    def set_age(self, age):
        if age > 0:
            self._age = age

# Usage
person = Person("John", 25)
print("Original Name:", person.get_name())
print("Original Age:", person.get_age())

person.set_name("Jane")
person.set_age(30)

print("Updated Name:", person.get_name())
print("Updated Age:", person.get_age())
```

In this example, we have a `Person` class with private attributes `_name` and `_age`. The getter methods (`get_name` and `get_age`) retrieve the values of these attributes, and the setter methods (`set_name` and `set_age`) modify the values.

The getter methods simply return the values of the corresponding attributes using the `self` keyword.

The setter methods allow for validation and conditional checks before modifying the attribute values. In this example, the `set_age` method checks if the age provided is a positive value before assigning it to the `_age` attribute.

In the usage section, we create an instance of the `Person` class named `person` with the name "John" and age 25. We can retrieve the name and age values using the getter methods `get_name` and `get_age`, respectively.

Then, we use the setter methods `set_name` and `set_age` to update the name to "Jane" and age to 30. Finally, we retrieve the updated values using the getter methods and print them.

Output:
```
Original Name: John
Original Age: 25
Updated Name: Jane
Updated Age: 30
```

By using getter and setter methods, we can control the access and modification of class attributes, enforcing validation rules or implementing additional logic as needed. This helps in maintaining data integrity and encapsulating the internal state of the class.

ans:5-Method overriding is a feature in object-oriented programming where a subclass provides a different implementation for a method that is already defined in its parent class. It allows the subclass to redefine the behavior of an inherited method based on its specific needs.

Here's an example code that demonstrates method overriding in Python:

```python
class Animal:
    def make_sound(self):
        print("Animal makes a sound.")


class Cat(Animal):
    def make_sound(self):
        print("Cat meows.")


class Dog(Animal):
    def make_sound(self):
        print("Dog barks.")


# Usage
animal = Animal()
animal.make_sound()

cat = Cat()
cat.make_sound()

dog = Dog()
dog.make_sound()
```

In this example, we have a base class `Animal` with a method called `make_sound` that prints a generic sound made by an animal. The `Cat` and `Dog` classes are subclasses of `Animal` and inherit the `make_sound` method.

However, both `Cat` and `Dog` classes override the `make_sound` method with their own implementation. The `Cat` class defines a `make_sound` method that prints "Cat meows," while the `Dog` class defines a `make_sound` method that prints "Dog barks."

When we create instances of the `Animal`, `Cat`, and `Dog` classes and call the `make_sound` method on each instance, the appropriate implementation of the method from the corresponding class is executed.

Output:
```
Animal makes a sound.
Cat meows.
Dog barks.
```

By overriding a method in a subclass, we can modify the behavior of that method to suit the specific needs and characteristics of the subclass. This allows for flexibility and customization in the behavior of objects while maintaining a common interface defined by the superclass.
